# Hangman ML Hackathon - Setup

This notebook sets up the environment and loads the data.


```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict, Counter
import random
import os
import sys

# Add src directory to path
sys.path.append('../src')

# Set random seeds for reproducibility
np.random.seed(42)
random.seed(42)

# Display settings
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
%matplotlib inline
```

```python
# Load the corpus
corpus_path = '../Data/corpus.txt'
test_path = '../Data/test.txt'

print("Loading corpus...")
with open(corpus_path, 'r') as f:
    corpus = [line.strip().lower() for line in f if line.strip()]

print(f"Loaded {len(corpus)} words from corpus")
print(f"Sample words: {corpus[:10]}")

# Load test set
print("\nLoading test set...")
with open(test_path, 'r') as f:
    test_words = [line.strip().lower() for line in f if line.strip()]

print(f"Loaded {len(test_words)} words from test set")
print(f"Sample words: {test_words[:10]}")
```
